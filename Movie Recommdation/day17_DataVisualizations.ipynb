# Types of Graphs 
1. plot -   line graph, between two parameters
2. bar  -   bar graph of single parameter (optional - cat_data)
3. hist -   histgram of bunch of data(Bunch of value occurence) i.e. frequency of single parameter
4. scatter - dot graph between two parameters
5. boxplot/box - statstical overview in form of box for a single parameter (optional - categorical data) 
                    it can be mean, outliers, qunatiles (25%, 75%), IQR
6. Stem - its more like bar graph without lines insteads of bars
7. violinplot - Violin shaped statstical overview graph
                spread of data (more like histogram)
                single parameter (optional cat_data)

```python
y2  = x**4
```

# Comparing graphs

```python
# Normal man way
plt.plot(x, y2)
plt.show()
plt.plot(x, y)
plt.show()
```

```python
#Smart Man Way
plt.plot(x, y, x, y2)
plt.legend(['y=x**3', "y2=x**4"])
plt.show()
```

```python
#Inteliigent Man Way
plt.plot(x, y, lw = 3, marker='o', markersize=15, markerfacecolor='yellow')
plt.plot(x, y2, marker='v', markersize = 10)
plt.legend(['y=x**3', "y2=x**4"])
plt.grid()
plt.show()
```

```python
#How legends will plot the graph
fig = plt.figure(figsize = (10,5))
plt.plot(x, y, lw = 3, marker='o', markersize=15, markerfacecolor='yellow')
plt.plot(x, y2, marker='v', markersize = 10)
plt.legend(['y=x**3', "y2=x**4"])
plt.grid()
plt.show()
```
