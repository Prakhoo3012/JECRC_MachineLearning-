# Graph Functions
1. plot                 -- Simple Line Plot                    
2. plot.area            -- Filled Version of Line Plot
3. plot.bar/ plot.barh  -- Bar Graph
4. plot.box             -- Box Plot
5. plot.density         -- Density Line (Smooth Curved)
6. plot.hexbin          -- Hexagon Version of scatter with Density
7. plot.hist            -- Histogram
8. plot.kde             -- Kernel Density EStimator (Density Graph)
9. plot.pie            -- Pie chart of Categorical Data
10. plot.scatter        -- Scatter Graph

# Types / Categories
1. Categorical PLots ---> (hue can be used in evrery function | comparison with categorical data)
    a. countplot (single parameter)
    b. barplot (one categorical, one numerical, calculates mean by default)
    c. boxplot
    d. violinplot
    e. swarmplot
2. Distribution Plots --->
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
3. Matrix Plots --->
    a. heatmap
    b. clustermap
    c. pairplot (same as above one)
4. Regression Plot --->
    a. lmplot
    b. regplot
5. Grid Plots  --->
    a. PairGrid
    b. JointGrid
    c. FacetGrid

# Bar Plot 
One parameter must be categorical and other should be numerical

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
```

```python
sns.barplot(x='day', y='total_bill', data=tips, hue='time', estimator=np.mean)
```

# Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time')
```

# Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='time')
```

# Swarm Plot

```python
fig = plt.figure(figsize = (10,5))
sns.swarmplot(x='day', y='total_bill', data=tips, hue='time')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker'); plt.show()
plt.figure(figsize = (10,5))
sns.swarmplot(x='day', y='total_bill', data=tips, hue='smoker'); plt.show()
plt.figure(figsize = (10,5))
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker'); plt.show()
```
